```python
# 代码实现节点
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

#define four nodes
a = Node(86)
b = Node(19)
c = Node(4)
d = Node(12)
# link the four nodes
a.next = b
b.next = c
c.next = d
head = a

# output the nodes
while head is not None:
    print(head.data)
    head = head.next
```

```python
# linked list
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = Node
    def is_empty():
        '''
        判断链表是否为空
        '''
        return self.head is None
    def append(self, data):
        node = Node(data)
        if self.head is None:
            self.head = node
            self.tail = node
        else:
            self.tail.next = node
            self.tail = node
    def iterate(self):# 遍历元素
        if not self.head:
             return 
        current = self.head
        yield current.data
        while current.next:
            current = current.next
            yield current.data
    def insert(self, index, value):
        '''在链表的指定索引插入一个新的节点'''
        current = self.head
        current_index = 0
        if self.head is None:
            raise Exception("The list is an empty list")
        while current_index < index - 1:
            current = current.next
            if current is None:
                raise Exception('......')
            current_index += 1
        node = Node(value)
        node.next = current.next # node的指针指向原节点的下一节点
        current.next = node # 原节点的下一节点指向node
        if node.next is None:
            self.tail = node

    def remove(self, index):
        '''移除链表指定索引元素'''
        current = self.head
        current_index = 0

        if self.head is None:
            raise Exception('The list is an empty list')
        while current_index < index - 1:
            current = current.next
            if current is None:
                raise Exception('......')
            current_index += 1
        if index == 0:
            self.head = current.next
            current = current.next
            return

        if self.head is self.tail:
            self.head = None
            self.tail = None
            return
        
        current.next = current.next.next
        if current.next is None:
            self.tail = current

    def size(self):
        '''返回链表长度'''
        current = self.head
        count = 0
        if current is None:
            return 'the linked list is empty'
        while current is not None:
            count += 1
            current = current.next
        return count
    
'''liked list easy operation'''
if __name__ == '__main__':
    l = LinkedList()
    for i in range(10):
        l.append(i)
    print(list(l.iterate()))
    l.insert(2, 10)
    print(list(l.iterate()))
    l.remove(0)
    print(list(l.iterate()))
    print(l.size())
```

```python
# two sum
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def addTwoNumbers(self,l1,l2):
        val1, val2 = [l1.val],[l2.val]
        while l1.next:
            val1.append(l1.next.val)
            l1 = l1.next

        while l2.next:
            val2.append(l2.next.val)
            l2 = l2.next
        
        num_1 = ''.join([str(i) for i in val1[::-1]])
        num_2 = ''.join([str(i) for i in val2[::-1]])

        sums = str(int(num_1)+int(num_2))[::-1]

        res = head = ListNode(int(sums[0]))

        for i in range(1, len(sums)):
            head.next = ListNode(sums[i])
            head = head.next
        return res
```
